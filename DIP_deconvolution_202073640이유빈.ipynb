```python
from skimage import data, restoration
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
astro = data.astronaut()

plt.imshow(astro)
plt.axis('off')
plt.show()
```

```python
astro_norm = (astro - astro.min()) / (astro.max()-astro.min())
```

```python
noise = np.random.normal(loc=0, scale=0.1, size=astro.shape)
```

```python
astro_noise = cv2.add(astro_norm, noise, dtype = cv2.CV_32F)
```

```python
plt.imshow(astro_noise)
plt.axis('off')
plt.show()
```

```python
psf_size = 5
psf = np.ones((psf_size, psf_size, 1)) / (psf_size**2)
```

```python
astro_noise[astro_noise<0.] = 0.
astro_noise[astro_noise>1.] = 1.
```

```python
deconvolved_RL = restoration.richardson_lucy(astro_noise, psf, iterations=30)
```

```python
plt.imshow(deconvolved_RL)
plt.axis('off')
plt.show()
```
